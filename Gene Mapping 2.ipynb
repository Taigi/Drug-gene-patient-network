```python
import pandas as pd
```

```python
db_data = pd.read_csv("uniprotlinks.csv")
```

```python
col = db_data["UniProt ID"]
```

```python
import mygene
```

```python
mg = mygene.MyGeneInfo()
```

```python
out = mg.querymany(col, scopes='uniprot', fields='symbol', species='human', returnall=True, as_dataframe=True)
```
