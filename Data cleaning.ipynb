```python
data_files = [
    "uniprotlinks.csv",
    "uniprot-all2.csv"
]

```

```python
data = {}
```

```python
for f in data_files:
    d = pd.read_csv("data/{0}".format(f))
    key_name = f.replace(".csv", "")
    data[key_name] = d 
```
