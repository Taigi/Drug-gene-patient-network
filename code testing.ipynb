```python
import pandas as pd
drugbankdata = pd.read_csv("uniprot links.csv")
column = drugbankdata["UniProt ID"]

##using mygenee module to map uniprot id's to their corresponding gene symbols. 
import mygene as mge
mg = mge.MyGeneInfo()

##calling querymany method to specify parameters: input=uniprot, output=symbol, returning all duplicates and as a dataframe. 

out = mg.querymany(column, scopes='uniprot', fields='symbol', species='human', returnall=True, as_dataframe=True)

```
